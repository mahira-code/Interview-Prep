## Problem Summary: Array Sum (HackerRank Style)

You are given an array of integers. Some numbers may be very large. task is to print the sum of all the elements in the array.
### Input Format:
N
x1,x2,....xn

### output Format:
Print a single integer — the sum of the array elements.

#### Constraints:
1 ≤ N ≤ 10

0 ≤ a[i] ≤ 10^10 

#### Sample Input
5
1000000001 1000000002 1000000003 1000000004 1000000005

#### Sample Output
1000000015



#### Option1 : using build in functions

```python
# Number of Array element input 
N = int(input('enter the number of array'))
#Array Element
x = list(map(int,input('enter array element with space').split()))

#sum 
total = sum(x)
print(total)
```

###### Explanation:
int(input):  Reads the number of elements N.

list(map(int,input().split())): Takes N space-separated inputs and converts them to integers, stored in list x.

total = sum(x): Uses Python’s built-in sum() function to add all elements.

print(): Prints the result.

###### Advantages:
Clean and readable.

Uses Python’s efficient built-in functions (sum, map).

Compact: only 4 main lines.

Time Complexity: O(N)

Space Complexity: O(N)


##### Line by Line Complexity Explanation
| Line                                  | Description                         | Time Complexity | Space Complexity |
| ------------------------------------- | ----------------------------------- | --------------- | ---------------- |
| `N = int(input())`                    | Reads number of elements            | **O(1)**        | **O(1)**         |
| `x = list(map(int, input().split()))` | Reads `N` elements, converts to int | **O(N)**        | **O(N)**         |
| `sum(x)`                              | Adds all elements                   | **O(N)**        | **O(1)**         |
| `print(total)`                        | Prints result                       | **O(1)**        | **O(1)**         |

##### Final Complexity 


| Metric             | Value    |
| ------------------ | -------- |
| Time Complexity  | **O(N)** |
| Space Complexity | **O(N)** |


#### Option2

```python
N = input("Enter number of Element in Array")

x = list(map(int,input("enter array element with space").split()))
sum = 0
for x1 in x:
    sum = sum+x1

print("Sum of Array Element:",sum)
```

Explanation:

Line 1: Prompts user for the number of elements (but doesn't convert it to int or use it later).

Line 2: Reads space-separated input and converts to integers.

Line 3-4: Uses a for loop to iterate through each element and calculate sum manually.

Line 5: Prints the final sum with a message.

Issues:

N is unused and not typecasted to int – this may be confusing or error-prone.

The variable sum is a built-in function in Python – shadowing it is not recommended.

More lines and manual looping make it slightly less efficient.

Advantages:
Good for beginners to understand loops and summation.

Explicit iteration might help understand logic flow.

##### Lineby Line Complexity Explanation:
| Line                 | Code                   | Time Complexity | Explanation                                            |
| -------------------- | ---------------------- | --------------- | ------------------------------------------------------ |
| `input(...)`         | Reading input          | **O(1)**        | Reading `N`, single input line.                        |
| `input(...).split()` | Split string into list | **O(N)**        | Splits a string of N numbers into N strings.           |
| `map(int, ...)`      | Convert each to int    | **O(N)**        | Each string to integer conversion takes constant time. |
| `list(...)`          | Collect mapped values  | **O(N)**        | Makes a list of N integers.                            |
| `for x1 in x:`       | Loop over array        | **O(N)**        | One loop for all N elements to compute the sum.        |
| `print(...)`         | Output result          | **O(1)**        | Just printing one line.                                |

##### Complexity Summary
